```python
# I tried to get a little fancy with a random generator referenced from subtstack - it had Javascript initially, but I researched and fiddled with google help to get it python ready
import random
#imports random library
def getRandomArbitrary(min_val, max_val):
    #"""Returns a random floating-point number between min (inclusive) and max (exclusive)."""
    return random.uniform(min_val, max_val)

def getRandomInt(min_val, max_val):
    #"""Returns a random integer between min (inclusive) and max (inclusive)."""
    return random.randint(min_val, max_val)

# Example Usage:
random_float = getRandomArbitrary(0, 100)
random_integer = getRandomInt(0, 99)

print(f"Random float: {random_float}")
print(f"Random integer: {random_integer}")

#Example of using the random numbers in the list assignment.
list_1 = []
for i in range(3):
    list_1.append(getRandomInt(0,99))

print(f"my list with 3 random integers:{list_1}")

list_2= []
for i in range(20):
    list_2.append(getRandomInt(0,99))

print(f"my second list with 7 random integers: {list_2}")

```
